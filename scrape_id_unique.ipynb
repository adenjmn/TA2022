```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

import pandas as pd
import requests
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import NoSuchElementException
import time

from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome('/home/ade/Documents/TA2022/chromedriver')
```

```python
data = pd.read_csv('id_unik.csv')
data.head()
```

```python
data.id.value_counts()
```

```python
# uniqueID = data.id.drop_duplicates()
uniqueID = data.drop_duplicates(subset=['id'])

uniqueID.id.value_counts()
```

```python
import re
import numpy as np

unikList = uniqueID['id'].tolist()
clean_uniqueID = []
for i in unikList:
    n_data = re.sub("[^0-9]", "", i)
    clean_uniqueID.append(n_data)

uniqueID['id_user'] = clean_uniqueID
uniqueID.head()
```

```python
name =[]
img_src =[]
points =[]
level =[]
n_reviews =[]
n_photos =[]
view_photos =[]
```

```python
df = uniqueID[2000:3000]
df
```

```python
def scrape():
    for i in df.id:
        url = i
        driver.get(url)

#         scrape poin
        try:
            in_point = driver.find_element("xpath",'//*[@id="uc9cqe-n1UuX-XUufyc-Wt0bsf-hUbt4d"]/span[3]/jsl').text
            points.append(in_point)
        except:
            points.append("NaN")
        
        print(len(points))
        
#         scrape level
        try:
            in_level = driver.find_element("xpath",'//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div[2]/div[3]/span').text
            level.append(in_level)
        except:
            level.append("NaN")
            
#         scrape n_review
        try:
            in_n_reviews = driver.find_element("xpath",'//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[6]/div[1]/div[2]/span/span[2]').text
            n_reviews.append(in_n_reviews)
        except:
            n_reviews.append("NaN")
        
#         otw img
        try:
            select = driver.find_element("xpath",'/html/body/div[3]/div[9]/div[9]/div/div/div[1]/div[2]/div/div[1]/div/div/div[5]/div/div/button[2]/div/div[1]')
            select.click()
#             time.sleep(2)
        except:
            pass
        
#         scrape photos
        try:
            in_n_photos = driver.find_element("xpath",'/html/body/div[3]/div[9]/div[9]/div/div/div[1]/div[2]/div/div[1]/div/div/div[6]/div[1]/div[2]/span[1]/span[2]').text
            n_photos.append(in_n_photos)
        except:
            n_photos.append("NaN")
        
#         scrape view img
        try:
            in_view_photos = driver.find_element("xpath",'/html/body/div[3]/div[9]/div[9]/div/div/div[1]/div[2]/div/div[1]/div/div/div[6]/div[1]/div[2]/span[2]/span[2]').text
            view_photos.append(in_view_photos)
        except:
            view_photos.append("NaN")
            
#         scrape img profil
        try:
            get_img = driver.find_element("xpath",'/html/body/div[3]/div[9]/div[9]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div[2]/div[1]/div/img').get_attribute("src")
            img_src.append(get_img)
        except:
            img_src.append("NaN")
            
    print("Total scrape",len(points))
```

```python
scrape()
```
